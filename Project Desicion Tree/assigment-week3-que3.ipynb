```python
q3=np.percentile(df['popularity'],75)
q1=np.percentile(df['popularity'],25)
q3,q1
```

```python
iqr=q3-q1
iqr
```

```python
ub=q3+(1.5*iqr)
lb=q1-(1.5*iqr)
ub,lb
```

```python
sns.boxplot(df['popularity']);
plt.title('outliers in popularity');
```

```python
plt.scatter(df['brand_name'],df['popularity'],c=df['best_price']);
```

```python
brand_name=['ALCATEL','Nokia','Apple','HUAWEI','Sigma mobile']
popularity=[422,1047,1101,1174,952]
plt.title('Bar Plot For Phones')
plt.xlabel('brand_name')
plt.ylabel('popularity')
plt.bar(brand_name,popularity,color='lavender',edgecolor='black');
plt.grid()
```
